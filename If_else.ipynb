```python
################ Parte 1 - Atividade 1 ################
idade = int(input("Digite sua idade: "))

if idade >= 18:
  print("Você é maior de idade.")
else:
  print("Você é menor de idade.")
```

```python
################ Atividade 2 ################
idade = int(input("Digite sua idade: "))

if idade <= 16:
  print("Você é menor de idade")
elif idade >= 16 and idade < 18:
  print("Voto opcional")
elif idade >= 18 and idade <= 59:
  print("Voto obrigatorio")
elif idade >= 60:
  print("Voto opcional")
else:
  print("Opção invalida")
```

```python
################ Atividade 3 ################
entrega = input("Digite o tipo de entrega: ")
if entrega == "padrão":
  print("o valor da entrega é R$10,00")
elif entrega == "expresso":
  print("O valor da entrega é R$20,00")
else:
  print("Opção de entrega invalida")
```

```python
################ Parte 2 - Atividade 1 ################
num = float(input("Digite o numero: "))
if num > 0:
  print("Numero positivo")
elif num == 0:
  print("Numero nulo")
else:
  print("Numero negativo")
```

```python
################ Atividade 2 ################
num = int(input("Digite seu numero: "))
if num % 2 == 0:
  print("Número par")
else:
  print("Número impar")
```

```python
################ Atividade 3 ################
notas = float(input("Digite sua nota : "))
if notas >= 7:
  print("Aprovado")
elif notas >= 5:
  print("Recuperação")
else:
  print("Reprovado")
```

```python
################ Atividade 4 ################
usuario = input("Digite o nome do usuário: " )
senha = int(input("Digite a senha: "))
if usuario == "admin" and senha == 1234:
  print("Login realizado")
else:
  print("Usuário ou senha inválidos")
```
